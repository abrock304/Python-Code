### Create Region Variable

```python
# Create a 'region' column based on the 'state' column to determine if there are any insights to each region.
# Check for null state values
all_nan = all_secure[all_secure['state'].isnull() == True]
```

### Identify Regular Activity Customers

```python
# Cross region and spender columns to create a table
cross_tab_all = pd.crosstab(index=all_secure['region'],
                             columns=all_secure['spender'],
                             normalize="index")
```

```python
cross_tab_all
```

From the 'cross_tab_all' table, we can see that there is no difference between each region when it comes to high- or low-spenders.

```python
# Create an exclusion flag for low-activity customers (less than 5 orders) and then exclude them from the data
all_secure.loc[all_secure['max_order'] < 5, 'activity']= 'Low activity'
```

```python
all_secure.loc[all_secure['max_order'] >= 5, 'activity']= 'Regular activity'
```

```python
all_secure['activity'].value_counts()
```

```python
low_activity = all_secure.loc[all_secure['activity'] == 'Low activity']
```

```python
low_activity.head()
```

```python
regular_activity = all_secure.loc[all_secure['activity'] != 'Low activity']
```

```python
regular_activity['activity'].value_counts()
```

```python
regular_activity.shape
```

```python
# Assign df to dataframe for easy use.
df = regular_activity
```

```python
# Drop 'activity' column since all entries are now considered 'regular activity'
df = df.drop(columns = ['activity'])
```

```python
df['avg_price'] = df.groupby(['user_id'])['prices'].transform(np.mean)
```
