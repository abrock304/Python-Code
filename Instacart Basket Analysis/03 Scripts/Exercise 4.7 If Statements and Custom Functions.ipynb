## Define Function

```python
# Define the function 'price_label'
def price_label(row):
    if row['prices'] <= 5:
        return 'Low-range product'
    elif (row['prices'] > 5) and (row['prices']<= 15):
        return 'Mid-range product'
    elif row['prices'] > 15:
        return 'High range'
    else: return 'Not enough data'
```

```python
# Apply function to merged dataframe
orders_products_merged['price_range'] = orders_products_merged.apply(price_label, axis =1)
```

```python
# Check the values for 'order_hour_of_day' to determine the desired time periods

orders_products_merged['order_hour_of_day'].value_counts(dropna = False)
```

We can divide these into 3 sections to describe the Most, Average, and Fewest order time periods. Most orders from 9am-5pm, Average orders from 7-9am and 5-11pm, and Fewest from 11pm-7am. 

```python
# Create a function that will label each time period.

hours = []

for x in orders_products_merged["order_hour_of_day"]:
  if (x >= 9) and (x <= 16):
    hours.append("Most orders") 
  elif ((x <= 6) and (x >= 0)) or (x==23):
    hours.append("Fewest orders")
  else:
    hours.append("Average orders")
```

```python
hours
```

```python
# Add busiest_time_period column to dataframe
orders_products_merged['busiest_time_periods'] = hours
```

```python
# Check the counts for each possible solution to busiest_time_periods
orders_products_merged['busiest_time_periods'].value_counts(dropna = False)
```
