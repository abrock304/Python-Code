```python
# Change date_joined column to datetime datatype.
customers = customers.astype({'date_joined': 'datetime64[ns]'})
```

```python
# Rename columns so format is consistent
customers = customers.rename(columns={'First Name':'first name', 'Surnam': 'last name', 'Gender': 'gender', 'STATE': 'state', 'Age': 'age'})
```

```python
# Check for duplicate values
dups = customers[customers.duplicated()]
dups
```

No duplicate entries in dataframe.

```python
# Check for mixed-data types
for col in customers.columns.tolist():
  weird = (customers[[col]].applymap(type) != customers[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (customers[weird]) > 0:
    print (col)
```

The column 'first name' has a mixed-data type. We will change this to a string data type.

```python
customers['first name'] = customers['first name'].astype('str')
```
